```python
#Measure of skewness of ‘bmi’, ‘age’ and ‘charges’ columns
skewness = pd.DataFrame({'skewness':[stats.skew(df.bmi),
                                     stats.skew(df.age),
                                     stats.skew(df.charges)]},
                        index = ['bmi','age','charges'])

skewness
```
